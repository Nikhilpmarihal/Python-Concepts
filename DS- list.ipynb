```python
# 4. Check if list is sorted
a = [12,13,15,16,17,23,24,25,36,78,90]
for i in range(len(a)-1):
    if a[i] < a[i+1]:
        continue
    else:
        print("Your list is not sorted")
        break
else:
    print("Your list is sorted")
```

```python
# 5. Left rotation by 1
a =[10,20,30,40,50]
for i in range(len(a)-1):
    a[i] , a[i+1] = a[i+1] , a[i]
print(a)
```

```python
# 6.Left rotation by k
k = int(input("How many times you want to rotate:"))
a =[10,20,30,40,50]
for i in range(k):
    for i in range(len(a)-1):
        a[i],a[i+1]  = a[i+1],a[i]
print(a)
```

```python
# 7. Reverse list 
a = [10,20,30,40,50]
b=len(a)-1
for i in range(len(a)//2):
    a[i],a[b] = a[b],a[i]
    b = b-1
print(a)
```

```python
# 8. Linear Search
a = [23,67,123,1,54,7,98,45,23,13,6,68]
search = 6
for i in range(len(a)):
    if a[i] == search:
        print(f"Element found at index {i}")
        break
else:
    print("Sorry no such element exist")
```

```python
#11. selection sort
a = [56,345,67,89,245,23,45,67,90,12,369]
for i in range(len(a)-1):
    j = i+1
    min =1
    for k in range(j,len(a)):
        if a[k] < a[min]:
            min =k
    a[i],a[min] = a[min],a[i]
print(a)
```
