```python
# 2.Greet by Gender(M/F) with gender case sensitive handling
gen = input("Please tell your gender in character (m,f):")
if gen == "m" or gen == "M" :
    print("Hello sir how are you")
elif gen == "f" or gen == "F":
    print("Hello Ma'am how are you")
else:
    print("Wrong Input , only provide (M  or F)")
```

```python
# 5. Day Number to Day Name
a = int(input("Enter the day(1-7):"))
if a == 1:
    print("Today is Monday")
elif a == 2:
    print("Today is Tuesday")
elif a == 3:
    print("Today is Wednesday")
elif a == 4:
    print("Today is Thursday")
elif a == 5:
    print("Today is Friday")
elif a == 6:
    print("Today is Saturday")
elif a == 7:
    print("Today is Sunday")
else:
    print("Input Error , Choose between (1-7) only")
```

```python
# 6. Greatest of Three Numbers
a = int(input("Enter the first number:"))
b = int(input("Enter the second number:"))
c = int(input("Enter the third number:"))
if a == b and b == c:
    print("All the numbers are equal")
elif a == b or b == c or c == a:
    print("Any two numbers are equal")
elif a > b and a > c:
    print(f"{a} is the greatest number")
elif b > a and b > c:
    print(f"{b} is the greatest number")
else:
    print(f"{c} is the greatest number")
```

```python
# 7.Leap Year Cheacker
year = int(input("Enter the year:"))
if year % 100 == 0 and year % 400 == 0:
    print("it is a leap year")
elif year % 100 != 0 and  year % 4 == 0:
    print("It is a leap year")
else:
    print("its not a leap year")
```

```python
# 8. Shop discount calculator
bill = int(input("Enter the total amount:"))
if bill >= 1000 and bill <= 4999 :
    print(f"You got a discount of 10%. Your final amount is {(bill * 90/100)}")
elif bill >= 5000:
    print(f"You got a discount of 20%. Your final amount is {(bill * 80/100)}")
else:
    print("No discount Available")
```
