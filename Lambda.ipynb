```python
#lambda expression

square = lambda a : print(a**2)
square(12)
add = lambda x,y : x+y
print(add(12,12))
```

```python
# map
# Purpose: Apply a function to every item of an iterable and return a new iterable

# Syntax: map(function, iterable)

def square(x):
    return x ** 2

a = [1, 2, 3, 4]
l = map(square, a)
print(list(l))
```

```python
#Filter
#purpose - Filter items from an iterable boased on a condition

#syntax - filter(function,iterable)

a = [1,2,3,4,5,6]

l = filter(lambda x : x%2 == 0,a)

print(list(l))
```

```python
#zip
#purpose : Combine multiple iterables into pairs of elements.

#syntax = zip(iterable 1,iterable2, ....)

name = ["Nik","Mik","Rik"]
ages = [24,22,23]

comb = zip(name,ages)

print(dict(list(comb)))


a = [1,2,3,4,5,6,7,8,9]

l = [i for i in a if i % 2 == 0]

print(l)

a = [1,2,3,4,5,6,7,8,9]

l = {i for i in a if i %2 ==0}

print(l)

a = [1,2,3,4,5,6,7,8,9]

l = {i:i**2 for i in a if i%2 ==0}

print(l)
```

```python
# Generators in Python
# Purpose: Generate values one by one instead of storing all values in memory

def my_generator():
    for i in range(5):
        yield i

gen = my_generator()
print(next(gen))
print(next(gen))
print(list(gen))   # remaining values


# Generator expression
sequence = (x**2 for x in range(5))

print(next(sequence))
print(next(sequence))
print(next(sequence))


# Decorators in Python
from functools import wraps

def my_decorator(func):
    @wraps(func)
    def wrapper():
        print("Hello, I will print before")
        func()
        print("Hello, I will print after")
    return wrapper

@my_decorator
def say_hello():
    print("Hello")

say_hello()


# Decorator with arguments
def decorate(func):
    @wraps(func)
    def wrapper(a, b):
        print("Your 2 numbers addition is:")
        result = func(a, b)
        print("Thank you for using us")
        return result
    return wrapper

@decorate
def addition(a, b):
    print(a + b)

addition(12, 12)
```
