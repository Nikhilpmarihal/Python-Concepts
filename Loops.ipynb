***WHILE LOOP***

```python
# 1.Print each digit in reverse order
a = int(input("Please enter the number"))
while a > 0:
    print(a%10)
    a = a//10
```

```python
# 2. sum of digits
a = int(input("Please enter the number"))
s = 0
while a > 0:
    s = s+ a% 10
    a = a//10
print(f"Your digits sum is {s}")
```

```python
# 3. Reverse a number
a = int(input("Please enter the number"))
rev = 0
while a > 0:
    rev = rev * 10 + a%10
    a = a//10
print(f"your number reverse is {rev}")
```

```python
# 4. Pallindromic number
a = int(input("Please enter the number"))
copy = a
rev = 0
while a > 0:
    rev = rev * 10 + a%10
    a = a//10
if rev == copy:
    print("Yes your number is pallindrome")
else:
    print("Yes your number is not pallindrome")
```

```python
# 5.Autpmorphic Number
a = int(input("Enter the number:"))
dup = a
square = a ** 2
count = 0
while a> 0:
    count = count + 1
    a = a//10
extract = square % (10**count)
if extract == dup:
    print("your number is automorphic")
else:
    print("Not automorphic number")
```
