```python
def pallindrome(x):
    rev = 0
    copy = x
    while x > 0:
        rev = rev * 10 + (x % 10)
        x = x // 10
    if rev == copy:
        return True
    else:
        return False 
print(pallindrome(121))
print(pallindrome(124))
```
